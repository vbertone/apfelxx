```python
import numpy as np
from sympy import Symbol, Eq, simplify
from sympy import sqrt, cosh, exp
from sympy.solvers import solve
from sympy import init_printing
init_printing()
```

```python
# Relevant variables
eta = Symbol(r'\eta')
x = Symbol('x')

# Kinematics
Q = Symbol('Q', positive=True)
qT = Symbol(r'|\mathbf{q}_T|', positive=True)
y = Symbol('y')
M = Symbol('M') #sqrt(Q**2+qT**2)

# Cut parameters
etamax = Symbol(r'\eta_{\rm max}', positive=True)
pTmin = Symbol(r'p_{T,\rm min}', positive=True)

# Definitions
ch = Symbol('\cosh(\eta-y)') #cosh(eta - y)
pTo = Q**2 / 2 / ( M * ch - qT * x)
```

```python
# xpTmin1
f = Eq(pTo, pTmin)
solve(f, x)
```

```python
# xpTmin2
f = Eq(qT**2 + pTo * ( pTo - 2 * qT * x), pTmin**2)
solve(f, x)
```

```python
f = Eq( ( M * exp(y) - pTo * exp(eta) ) / ( M * exp(-y) - pTo * exp(-eta) ) , exp(2*etamax))
solve(f, x)
```

```python
che = Symbol('\cosh(\eta)')
she = sqrt(che**2 - 1)
f = Eq(( che * ( Q**2 - 2 * pTmin**2 ) - Q * sqrt( Q**2 * she**2 + pTmin**2) ) / ( Q**2 - pTmin**2 ), 0)
solve(f, Q)
```
